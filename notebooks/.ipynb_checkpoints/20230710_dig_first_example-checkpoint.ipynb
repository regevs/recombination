# Dig in an example

```python
def convert_quality_string_to_array(quality_string):
    offset = 33  # Offset value for ASCII characters in FASTQ format
    quality_array = np.frombuffer(quality_string.encode(), dtype=np.uint8)
    return quality_array - offset
```

```python
alignment_bam_file = Path("/lustre/scratch126/casm/team154pc/sl17/03.sperm/02.results/01.read_alignment/01.ccs/04.hifiasm/02.hifiasm_0.19.5-r592/02.chromosome_length_scaffolds") \
    / sample_name / f"{sample_name}.minimap2.primary_alignments.sorted.bam"
```

```python
from hapfusion import bamlib
import pysam

for aln in pysam.AlignmentFile(alignment_bam_file):
    if aln.query_name == read_name:
        ccs = hapfusion.bamlib.BAM(aln)
        ccs.cs2tpos2qbase()
        break
```

```python
hetsnp_positions_in_read_coords_0based = [ccs.rpos2qpos[x] for x in hetsnp_positions]
```

```python
# These should be the same
print(aln.query_length)
print(len(fo.getSeq()))
```

### Plot

```python
colors = []
alphas = []
sizes = []
for pos_0based in np.arange(aln.query_length):
    if pos_0based in hetsnp_positions_in_read_coords_0based:
        if np.array(list(ccs_hbit))[np.where(hetsnp_positions_in_read_coords_0based == pos_0based)[0][0]] == '0':
            colors.append("C1")
            alphas.append(1)
            sizes.append(10)
        else:
            colors.append("C2")
            alphas.append(1)
            sizes.append(10)
    else:
        colors.append("C0")
        alphas.append(1)
        sizes.append(0.01)

fig, ax = plt.subplots(figsize=(10,4))
ax.scatter(
    np.arange(aln.query_length),
    read_sequencing_qualities,
    alpha=alphas,
    c=np.array(colors),
    s=sizes,
)

plt.title("Calling quality scores at SNPs");
plt.xlabel("Position");
plt.ylabel("Base quality");
plt.legend(handles=[
    matplotlib.lines.Line2D([0],[0], color='C0', label='Hom', marker='o', markersize=1, linestyle=''),
    matplotlib.lines.Line2D([0],[0], color='C1', label='SNP like haplotype 0', marker='o', markersize=5, linestyle=''),
    matplotlib.lines.Line2D([0],[0], color='C2', label='SNP like haplotype 1', marker='o', markersize=5, linestyle=''),
])
```

## Check the read->assembly alignment

```python
# TODO: Change after version changes
# read_alignment_path = Path("/lustre/scratch126/casm/team154pc/sl17/03.sperm/02.results/01.read_alignment/01.ccs/04.hifiasm/01.hifiasm_0.16.1-r375/") \
#                           / sample_name / f"{sample_name}.minimap2_ccs.primary_alignments.sorted.bam"
read_alignment_path = Path("/lustre/scratch126/casm/team154pc/sl17/03.sperm/02.results/01.read_alignment/01.ccs/04.hifiasm/02.hifiasm_0.19.5-r592/02.chromosome_length_scaffolds") \
    / sample_name / f"{sample_name}.minimap2.primary_alignments.sorted.bam"
```

```python
for read_aln in pysam.AlignmentFile(read_alignment_path):
    if read_aln.query_name == read_name:
        read_ccs = hapfusion.bamlib.BAM(read_aln)
        read_ccs.cs2tpos2qbase()
        break
```

```python
def get_offset_from_diagonal(aligned_pairs):
    p = 0
    cur_read_pos, cur_ref_pos = aligned_pairs[p]
    while cur_ref_pos is None:
        p += 1
        cur_read_pos, cur_ref_pos = aligned_pairs[p]

    diagonal_ref_pos = (cur_ref_pos - cur_read_pos)
    offsets = [0]
    for read_pos, ref_pos in aligned_pairs[p+1:]:
        if read_pos is None:
            read_pos = cur_read_pos
        if ref_pos is None:
            continue
            
        offsets.append((ref_pos - read_pos) - diagonal_ref_pos)
        cur_read_pos, cur_ref_pos = read_pos, ref_pos
    return np.array(offsets)
```

```python
mapping_delta = get_offset_from_diagonal(read_aln.get_aligned_pairs())
```

```python
colors = []
alphas = []
sizes = []
for ref_pos_0based in np.arange(read_aln.reference_start, read_aln.reference_end):
    if ref_pos_0based+1 in hetsnp_positions:
        if np.array(list(ccs_hbit))[np.where(hetsnp_positions == ref_pos_0based+1)[0][0]] == '0':
            colors.append("C1")
            alphas.append(1)
            sizes.append(10)
        else:
            colors.append("C2")
            alphas.append(1)
            sizes.append(10)
    else:
        colors.append("C0")
        alphas.append(1)
        sizes.append(0.01)

fig, ax = plt.subplots(figsize=(10,4))
ax.scatter(
    np.arange(read_aln.reference_start, read_aln.reference_end),
    mapping_delta,
    alpha=alphas,
    c=np.array(colors),
    s=sizes,
)

plt.title("Alignment offset from diagonal");
plt.xlabel("Position");
plt.ylabel("Relative position");
plt.legend(handles=[
    matplotlib.lines.Line2D([0],[0], color='C0', label='Hom', marker='o', markersize=1, linestyle=''),
    matplotlib.lines.Line2D([0],[0], color='C1', label='SNP like haplotype 0', marker='o', markersize=5, linestyle=''),
    matplotlib.lines.Line2D([0],[0], color='C2', label='SNP like haplotype 1', marker='o', markersize=5, linestyle=''),
])
```

### Check other reads mapping to that same region

```python
fig, ax = plt.subplots(figsize=(10,4))

for other_read in pysam.AlignmentFile(read_alignment_path).fetch(chrom_name + "_RagTag", read_aln.reference_start, read_aln.reference_end):
    ax.plot(
        np.arange(other_read.reference_start, other_read.reference_end),
        get_offset_from_diagonal(other_read.get_aligned_pairs()),
        alpha=1,
        #s=0.5,
        lw=1,       
    )

ax.set_title("Alignment offset from diagonal");
ax.set_xlabel("Position");
ax.set_ylabel("Relative position");
ax.set_ylim(-200,200)
```

```python
fig, ax = plt.subplots(figsize=(8,5))

for other_read in pysam.AlignmentFile(read_alignment_path).fetch(chrom_name + "_RagTag", read_aln.reference_start, read_aln.reference_end):
    read_coords, ref_coords = np.array([(read_pos, ref_pos) for read_pos, ref_pos in other_read.get_aligned_pairs() \
              if read_pos is not None and ref_pos is not None]).T
    ax.plot(ref_coords, read_coords, lw=1, alpha=0.5);

read_coords, ref_coords = np.array([(read_pos, ref_pos) for read_pos, ref_pos in read_aln.get_aligned_pairs() \
              if read_pos is not None and ref_pos is not None]).T
ax.plot(ref_coords, read_coords, lw=3, alpha=1, color="C0");
ax.set_xlabel("Position along reference")
ax.set_ylabel("Position along read")
```

## Check the read coverage to see if contigs assembled well

```python
coverages_per_nuc = pysam.AlignmentFile(read_alignment_path).count_coverage(chrom_name + "_RagTag", start_pos, end_pos)
```

```python
colors = []
alphas = []
sizes = []
for ref_pos_1based in np.arange(start_pos, end_pos):
    if ref_pos_1based in hetsnp_positions:
        if np.array(list(ccs_hbit))[np.where(hetsnp_positions == ref_pos_1based)[0][0]] == '0':
            colors.append("C1")
            alphas.append(1)
            sizes.append(10)
        else:
            colors.append("C2")
            alphas.append(1)
            sizes.append(10)
    else:
        colors.append("C0")
        alphas.append(1)
        sizes.append(0.1)

fig, ax = plt.subplots(figsize=(10,4))
ax.scatter(
    np.arange(start_pos, end_pos),
    np.array(coverages_per_nuc).sum(axis=0),
    alpha=alphas,
    c=np.array(colors),
    s=sizes,
)

plt.title("Coverage");
plt.xlabel("Position along contig");
plt.ylabel("Coverage");
plt.legend(handles=[
    matplotlib.lines.Line2D([0],[0], color='C0', label='Hom', marker='o', markersize=1, linestyle=''),
    matplotlib.lines.Line2D([0],[0], color='C1', label='SNP like haplotype 0', marker='o', markersize=5, linestyle=''),
    matplotlib.lines.Line2D([0],[0], color='C2', label='SNP like haplotype 1', marker='o', markersize=5, linestyle=''),
])

plt.xlim(162500,162700)

```
